```python
# Handle missing data i.e, Age data & Cabin Data
sns.boxplot(x='Pclass',y='Age' ,data=df)
```

```python
# Populate Age Data

def impute_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 23
    else: 
        return Age
                 
    
```

```python
df['Age']=df[['Age','Pclass']].apply(impute_age,axis=1)
```
